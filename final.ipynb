```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

final = pd.read_csv('final.csv')
```

```python
df=final
df=df.dropna()
df.isnull().sum()
```

```python
befor_encoding=df
```

```python
from sklearn.preprocessing import LabelEncoder
map = LabelEncoder()
df['Map'] = map.fit_transform(df['Map'])

player= LabelEncoder()
df['PlayerName'] =player.fit_transform(df['PlayerName'])

agent=LabelEncoder()
df['Agent']=agent.fit_transform(df['Agent'])
```

```python
import pickle
with open('map_encoder.pkl', 'wb') as f:
    pickle.dump(map, f)
```

```python
with open('Playername_encoder.pkl', 'wb') as f:
    pickle.dump(player, f)
```

```python
with open('agent_encoder.pkl', 'wb') as f:
    pickle.dump(agent, f)
```

```python
test = pd.DataFrame(columns=['C1', 'C2'])
test['C1']= df[['PlayerTeam', 'EnemyTeam']].stack().reset_index(drop=True)
new=LabelEncoder()
test['C2']=new.fit_transform(test['C1'])
test=test.drop_duplicates()
mapping = dict(zip(test['C1'], test['C2']))

# Replace the values in the 'Category' column with the corresponding values from df2
df['PlayerTeam'] = df['PlayerTeam'].map(mapping)
df['EnemyTeam'] = df['EnemyTeam'].map(mapping)
df['Winner'] = df['Winner'].map(mapping)
```

```python
df['PlayerTeamWin'] = df.apply(lambda row: 1 if row['Winner'] == row['PlayerTeam'] else 0, axis=1)
df['EnemyTeamWin'] = df.apply(lambda row: 1 if row['Winner'] == row['EnemyTeam'] else 0, axis=1)
```

```python
df['KDA'] = (df['Kills'] + df['Assists']) / df['Deaths'].replace(0,1)
```

```python
X = df[['GameID', 'Map', 'PlayerName', 'Agent','ACS','KDA','PlayerTeamWin']]
Xv=X.iloc[:,:-1].values
Yv=X.iloc[:,-1].values
```

```python
df.head(10)
```

```python
df.to_csv('average.csv',index=False)
```

```python
scaler = StandardScaler()
Xv = scaler.fit_transform(Xv)
```

```python
with open('scaler_object.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xv, Yv, test_size=0.1, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
from sklearn.metrics import confusion_matrix, accuracy_score
y_p= classifier.predict(X_test)
cm = confusion_matrix(y_test, y_p)
print(cm)
accuracy_score(y_test, y_p)
classifier.predict_proba(y_p)
```
