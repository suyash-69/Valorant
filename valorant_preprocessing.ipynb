```python
import sqlite3
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
final = pd.read_csv('final.csv')
```

```python
final.head(10)
```

```python
final['EnemyTeam']=0
final['PlayerTeam']=0
final['EnemyTeamRounds']=0
final['PlayerTeamRounds']=0
def assign_teams(row):
    if row['Team1'][:3] == row['TeamAbbreviation'][:3] :
        row['PlayerTeam'] = row['Team1']
        row['EnemyTeam'] = row['Team2']
        row['EnemyTeamRounds']=row['Team2_TotalRounds']
        row['PlayerTeamRounds']=row['Team1_TotalRounds']
    else:
        row['PlayerTeam'] = row['Team2']
        row['EnemyTeam'] = row['Team1']
        row['EnemyTeamRounds']=row['Team1_TotalRounds']
        row['PlayerTeamRounds']=row['Team2_TotalRounds']
    return row

final = final.apply(assign_teams, axis=1)

def assign_winner(row):
    if row['PlayerTeamRounds'] >row['EnemyTeamRounds'] :
      row['Winner']=row['PlayerTeam']
        
    else:
      row['Winner']=row['EnemyTeam']
    return row
final = final.apply(assign_winner, axis=1)
```

```python
final.drop(['Team1ID', 'Team2ID','Team1_TotalRounds','Team2_TotalRounds','TeamAbbreviation','Team1','Team2','PlayerID'], axis='columns', inplace=True)
```
